## Loading .tsv files from Volume to Schema as tables

```python
## Replace table if already exists
spark.sql("DROP TABLE IF EXISTS imdb.raw.name_basics")
spark.sql("DROP TABLE IF EXISTS imdb.raw.title_akas")
spark.sql("DROP TABLE IF EXISTS imdb.raw.title_basics")
spark.sql("DROP TABLE IF EXISTS imdb.raw.title_principals")
spark.sql("DROP TABLE IF EXISTS imdb.raw.title_ratings")

## Loading all tables onto raw schema
tables = [
    ("name.basics.tsv",       "name_basics"),
    ("title.akas.tsv",      "title_akas"),
    ("title.basics.tsv",        "title_basics"),
    ("title.principals.tsv",   "title_principals"),
    ("title.ratings.tsv",         "title_ratings")
]

volume_path = "dbfs:/Volumes/imdb/raw/imdb_files/"

for file, table in tables:
    df = spark.read.csv(
        volume_path + file,
        sep="\t",
        header=True,
        nullValue="\\N",
        inferSchema=True
    )
    
    df.write.mode("overwrite").saveAsTable(f"imdb.raw.{table}")

    print(f"Loaded: {table}")
```
